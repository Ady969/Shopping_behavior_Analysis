```python
import pandas as pd

# 1. Load the data
# If you are running this in Google Colab, you can upload the file first:
# from google.colab import files
# uploaded = files.upload()
# file_path = 'shopping_behavior.csv'

# For this script, we assume the file name is 'shopping_behavior.csv'
file_path = 'customer_shopping_behavior_1.csv'

try:
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please ensure it is uploaded to Colab.")

# 2. Basic Info and Missing Values
print("--- Basic Information ---")
print(df.info())

print("\n--- Missing Values Check ---")
print(df.isnull().sum())

# 3. Key Metrics Calculations
# Total Revenue: Sum of the 'Purchase Amount (USD)' column
total_revenue = df['Purchase Amount (USD)'].sum()

# Average Purchase Amount: Mean of the 'Purchase Amount (USD)' column
avg_purchase = df['Purchase Amount (USD)'].mean()

# Total Unique Customers: Count of unique 'Customer ID'
unique_customers = df['Customer ID'].nunique()

# Displaying the results
print("\n--- Key Business Metrics ---")
print(f"Total Revenue:            ${total_revenue:,.2f}")
print(f"Average Purchase Amount:  ${avg_purchase:,.2f}")
print(f"Total Unique Customers:   {unique_customers}")

# Optional: Quick look at the first few rows
print("\n--- Dataset Preview ---")
print(df.head())
```

```python
# 1. Top 5 selling items by total Purchase Amount
top_5_items = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(5)

print("--- Top 5 Selling Items by Revenue ---")
print(top_5_items.apply(lambda x: f"${x:,.2f}"))

# 2. Average purchase amount and order count by Gender
gender_analysis = df.groupby('Gender')['Purchase Amount (USD)'].agg(['mean', 'count'])
gender_analysis.columns = ['Average Purchase Amount', 'Order Count']

print("\n--- Gender Analysis ---")
print(gender_analysis)

# 3. Customer distribution by Age groups
# Defining bins and labels
bins = [18, 30, 50, df['Age'].max()]
labels = ['18-30', split_31_50 := '31-50', '51+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)
age_distribution = df['Age Group'].value_counts().sort_index()

print("\n--- Customer Distribution by Age Group ---")
print(age_distribution)

# 4. Average Review Rating by Subscription Status
subscription_rating = df.groupby('Subscription Status')['Review Rating'].mean()

print("\n--- Average Review Rating by Subscription ---")
print(subscription_rating)
```
